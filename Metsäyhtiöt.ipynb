```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
upm=pd.read_csv('upm.csv', sep = ';', decimal = ',', skiprows=1)
storaenso=pd.read_csv('storaenso.csv', sep = ';', decimal = ',', skiprows=1)
```

```python
upm.tail()
```

```python
storaenso.tail()
```

```python
upm.index=pd.to_datetime(upm['Date'],dayfirst=True)
storaenso.index=pd.to_datetime(storaenso['Date'],dayfirst=True)
```

```python
upm.sort_index(inplace = True)
storaenso.sort_index(inplace = True)
```

```python
upm['Closing price'].plot()
```

```python
storaenso['Closing price'].plot()
```

```python
storaenso['Closing price']['2019-01'].plot()
```

```python
fig1, ax1 = plt.subplots(figsize = (10, 6))

color = 'C0'
ax1.set_ylabel('upm', color=color)
ax1.plot(upm['Closing price'], color=color)
ax1.tick_params(axis='y', labelcolor=color, grid_color=color)

ax2 = ax1.twinx()

color = 'C1'
ax2.set_ylabel('storaenso', color=color)
ax2.plot(storaenso['Closing price'], color=color)
ax2.tick_params(axis='y', labelcolor=color, grid_color=color)
```

```python
upm['Closing price'].plot()
upm['Closing price'].rolling(100).mean().plot()
```

```python
storaenso['Closing price'].plot()
storaenso['Closing price'].rolling(100).mean().plot()
```
