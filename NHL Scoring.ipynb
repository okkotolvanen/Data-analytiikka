```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.tsa.api import SimpleExpSmoothing, Holt, ExponentialSmoothing
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
series=[127, 96, 121, 96, 106 ,94 ,0.1 ,125, 120, 112, 113, 112, 104, 109, 60, 104, 87, 106, 100, 108]
index=pd.date_range('1999', periods=20, freq='AS-APR')
data=pd.DataFrame(series, index=index)
data.columns=['Points']
data
```

```python
data.plot()
```

```python
fit1=ExponentialSmoothing(data['Points'], seasonal_periods=4, trend='add', seasonal='mul').fit()
```

```python
data['Predict']=fit1.fittedvalues
data.plot()
```

```python
data['Resid']=fit1.resid
data['Resid'].plot()
```

```python
index=pd.date_range('2019-04-01', periods=4, freq='AS-APR')
datap=pd.DataFrame(fit1.forecast(4),index=index)
datap.columns=['Predict']
data['Points'].plot()
datap['Predict'].plot()
```
