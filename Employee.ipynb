```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('employee.csv')
df.head(100)
```

```python
df.shape
```

```python
#kuinka paljon tyÃ¶ntekijÃ¶itÃ¤ eri osastoilla
df1=pd.crosstab(df['Department'],'lkm')
df1.columns.name=''
df1
```

```python
#ja prosentit samasta
df1['%']=df1/df1.sum()
df1.style.format({'%':'{:.1%}'})
```

```python
#tyÃ¶ntekijÃ¶iden koulutustausta
df2=pd.crosstab(df['EducationField'],'lkm')
df2.columns.name=''
df2
```

```python
#koulutus vaakapylvÃ¤inÃ¤
df2.sort_values(by='lkm', ascending=False).plot.barh(legend=False, color='C6')
```

```python
#ikÃ¤luokat
bins=[20,30,40,50,60]
df['AgeGroup']=pd.cut(df['Age'],bins=bins)
df3=pd.crosstab(df['AgeGroup'],'lkm')
df3.columns.name=''
df3
```

```python
#ikÃ¤luokat, histogrammi
df['Age'].plot.hist(bins)
plt.xlabel('Age')
plt.ylabel('Amount')
```
